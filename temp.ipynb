```python
import os
import cv2
from PIL import Image
```

```python
def to_bmp(inpath: str, outpath: str):
    with Image.open(inpath) as img:
        # print(f'{img.mode}')
        mode = img.mode

        if img.mode == 'CMYK':
            img = img.convert('RGB')

        if img.size != (671, 471):
            img = img.resize((671, 471))

        print(f'Converted ({mode}) {inpath=} to: {outpath=}')
        img.save(outpath, bitmap_format="bmp")
to_bmp('/Users/mahmoudhamido/Downloads/99(1).bmp', '/Users/mahmoudhamido/Downloads/99(1).bmp')
```

```python
frompath = '/Users/mahmoudhamido/templates/images'
topath = '/Users/mahmoudhamido/Downloads/forgery-detection-neural-network/datasets/midv-2020'

for doctype in os.listdir(frompath):
    docpath = os.path.join(frompath, doctype)
    targetdir = os.path.join(topath, doctype, 'original_genuine')

    if os.path.isdir(targetdir):
        for file in os.listdir(targetdir):
            filepath = os.path.join(targetdir, file)
            os.remove(filepath)

    for imgname in os.listdir(docpath):
        imgpath = os.path.join(docpath, imgname)
        targetpath = os.path.join(targetdir, imgname.replace('.jpg', '.bmp'))
        
        print(f'{imgpath=}')
        print(f'{targetpath=}')
        to_bmp(imgpath, targetpath)

    # print(targetdir)
```
